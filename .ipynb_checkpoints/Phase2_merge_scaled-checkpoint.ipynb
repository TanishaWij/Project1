```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def erp_per_lga(crime, lga):
    erps = []
    erp_column = (crime.loc[(crime["reference_period"] == 2013)& (crime["Local_Government_Area"] == lga), ["LGA_ERP"]])
    erps = erp_column.values 
    return erps[0][0]

def crime_by_LGA(crime, lga):
    sum_crime = (crime.loc[(crime["reference_period"] == 2013) & (crime["Local_Government_Area"] == lga) , ["Offence_Count"]])
    summ = sum_crime.sum()
    return (summ[-1])

def num_schools_by_lga(edu,lga):
    ''' 
    Returns the sum of the Government schools in the specified LGA 
    using boolean indexing
    '''
    num = (edu.loc[(edu["Local_Government_Area"] == lga), \
            ["Local_Government_Area"]]).Local_Government_Area.value_counts()
    return (num[0])

def unique_lga(data):
    ''' 
    Calculates the unique LGAs 
    '''
    lga =[]  
    all_lgas = data.Local_Government_Area.value_counts()  
    for key,value in all_lgas.iteritems():
        lga.append(key)
    lga.sort()
    return (lga)
    
def coordinates(crime, edu):
    lgas= []
    sum_crimes = []
    sum_schools = []
    erps = []
    lgas = unique_lga(crime)
    
    for lga in lgas:
        
        lga_crime = crime_by_LGA(crime, lga)
        sum_crimes.append(lga_crime)
        lga_schools = num_schools_by_lga(edu, lga)
        sum_schools.append(lga_schools)
 
    return(sum_crimes, sum_schools)
    
def scatter_plot(crime,edu):
    #number of schools as the x coordianate and the crime count as the y
    offences = []
    num_schools = []
    offences, num_schools = coordinates(crime, edu)
    
    plt.plot( num_schools, offences, 'bo', color= (.502, 0, .502))
    plt.xlabel( "Number of schools", color= (.502, 0, .502), fontsize = 25)
    plt.ylabel( "Offence rate", color= (.502, 0, .502), fontsize = 25)
    plt.grid(True)
    plt.show()
    
       
    
def main(): 
    crime = pd.read_csv("crimebylocationdatatable.csv")  
    edu = pd.read_csv("all_school.csv")
    scatter_plot(crime,edu)
    

    

if __name__ == "__main__":
    main()
```
